```python
import pandas as pd

# 1. Create a Series from a dictionary of state abbreviations to population
state_populations = {
    "CA": 39_500_000,
    "TX": 29_000_000,
    "FL": 21_500_000,
    "NY": 19_500_000,
    "PA": 12_800_000
}

state_pop_series = pd.Series(state_populations)
print("State Populations Series:")
print(state_pop_series)
print(f"\nPopulation of Texas: {state_pop_series['TX']:,}")
print("\nPopulations in millions:")
print(state_pop_series / 1_000_000)

# 2. Create a DataFrame for 5 students
student_data = {
    "Name": ["Emma", "Liam", "Olivia", "Noah", "Ava"],
    "GPA": [3.8, 3.2, 3.9, 2.9, 3.7],
    "Credits": [105, 75, 115, 60, 95],
    "Graduating": [True, False, True, False, False]
}

students_df = pd.DataFrame(student_data)
print("\nStudent DataFrame:")
print(students_df)

# 3. Access a row using .loc[] and a column using bracket notation
print("\nOlivia's record (using .loc[2]):")
print(students_df.loc[2])

print("\nGPA column (using bracket notation):")
print(students_df["GPA"])

# 4. Print the .shape, .columns, and .info() of your DataFrame
print(f"\nDataFrame shape: {students_df.shape}")
print(f"DataFrame columns: {students_df.columns.tolist()}")

print("\nDataFrame info:")
print(students_df.info())

# ðŸŽ¯ Bonus: Rename a column just to mess with the future grader
students_df.rename(columns={"Credits": "Credit_Hours"}, inplace=True)
print("\nAfter renaming 'Credits' to 'Credit_Hours':")
print(students_df.head())
```
