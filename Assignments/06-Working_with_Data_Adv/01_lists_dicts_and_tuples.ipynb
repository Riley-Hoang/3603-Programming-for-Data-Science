```python
# 1. Create a list of top 5 favorite foods. Print the last item using a negative index.
favorite_foods = ["pizza", "sushi", "tacos", "pasta", "ice cream"]
print("Last favorite food:", favorite_foods[-1])  # ice cream

# 2. Create a dictionary that maps student names to GPA and add a new student.
students = {
    "David": 3.9,
    "Bob": 3.7,
    "Charlie": 3.8
}
print("Original students:", students)

students["Diana"] = 3.95  # Adding a new student
print("Updated students:", students)

# 3. Make a tuple of the days of the week + Try to change one + Document what happens.
days_of_week = ("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday")
print("Days of week:", days_of_week)

# Trying to change a tuple element:
try:
    days_of_week[0] = "Funday"
except TypeError as e:
    print(f"Error when trying to modify tuple: {e}")

# 4. Create a set of course names. Add a duplicate.
courses = {"Math", "Science", "History", "English"}
print("Original courses:", courses)

courses.add("Math")  # Adding a duplicate
print("After adding duplicate:", courses)

# Use a list comprehension to build a list of all squares from 1-20
squares = [x**2 for x in range(1, 21)]
print("Squares from 1-20:", squares)
```
