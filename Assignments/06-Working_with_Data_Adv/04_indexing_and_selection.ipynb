```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = '/content/sample_data/california_housing_train.csv'
df = pd.read_csv(file_path)
```

```python
# 1. Select the first row using .iloc[], then again using .loc[]
print("First row using .iloc[]:")
print(df.iloc[0])

print("\nFirst row using .loc[]:")
print(df.loc[0])
```

```python
# 2. Slice the first 5 rows and select only housing_median_age, median_income, and median_house_value
first_five = df.iloc[0:5][["housing_median_age", "median_income", "median_house_value"]]
print("First 5 rows with selected columns:")
display(first_five)
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=first_five)
```

```python
# 3. Filter the dataset for houses with more than 2000 total rooms
large_houses = df[df["total_rooms"] > 2000]
print("Houses with more than 2000 total rooms:")
display(large_houses[["total_rooms", "median_house_value", "population"]].head())
```

```python
# 4. Create a DataFrame of houses with median income between 3 and 5 and house value under $150K
middle_income_affordable = df[(df["median_income"] >= 3) & (df["median_income"] <= 5) & (df["median_house_value"] < 150000)]
print("Middle income affordable houses:")
display(middle_income_affordable[["median_income", "median_house_value", "total_bedrooms"]].head())
```

```python
# 5. Use .loc[] to select rows 1020 and only columns median_income, total_rooms, and population
rows_10_20 = df.loc[10:20, ["median_income", "total_rooms", "population"]]
print("Rows 10-20 with selected columns:")
display(rows_10_20)
```

```python
#  Bonus: Find houses with zero bedrooms and print only longitude, latitude, and total_bedrooms
zero_bedrooms = df[df["total_bedrooms"] == 0]
print("Houses with zero bedrooms:")
display(zero_bedrooms[["longitude", "latitude", "total_bedrooms"]].head())
```
