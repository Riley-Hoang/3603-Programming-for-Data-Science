```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
#penguins = sns.load_dataset("penguins")
penguins = pd.read_csv("/content/sample_data/penguins.csv")
penguins.head()
```

## 1. Explore Distributions

First, letâ€™s look at the distribution of **body mass**.

```python
sns.histplot(data=penguins, x="body_mass_g", kde=True, hue="species")
plt.title("Distribution of Body Mass by Species")
plt.show()
```

ðŸ‘‰ **Task:** Do some species have higher average body mass than others?

<p><b>Observation:</b></p>
<ul>
  <li><b>Gentoo penguins</b> have the <b>highest average body mass</b>.</li>
  <li><b>Chinstrap penguins</b> are intermediate.</li>
  <li><b>Adelie penguins</b> have the <b>lowest body mass</b>.</li>
</ul>


## 2. Relationships Between Variables

Scatterplots help us see trends between two quantitative variables.

```python
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g", hue="species")
plt.title("Flipper Length vs Body Mass")
plt.show()
```

ðŸ‘‰ **Task:** Which species tends to have longer flippers AND higher body mass?

<p><b>Observation:</b></p>
<ul>
  <li><b>Gentoo penguins</b> tend to have <b>longer flippers and higher body mass</b>.</li>
  <li><b>Adelie penguins</b> are smaller in both measures.</li>
  <li><b>Chinstrap penguins</b> are in between.</li>
</ul>


## 3. Box Plots for Outliers

Box plots are great for spotting outliers.

```python
sns.boxplot(data=penguins, x="species", y="bill_length_mm", palette="Set2")
plt.title("Bill Length by Species")
plt.show()
```

ðŸ‘‰ **Task:** Can you spot any outliers in the bill length data?

<p><b>Observation:</b></p>
<ul>
  <li>Outliers appear as <b>dots outside the whiskers</b>.</li>
  <li>Some <b>Gentoo penguins</b> have unusually long bills.</li>
  <li>Some <b>Adelie penguins</b> may have very short or very long bills.</li>
</ul>


## 4. Correlation Heatmap

Correlation heatmaps let us see overall variable relationships.

```python
corr = penguins.dropna().corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

ðŸ‘‰ **Task:** Which variables appear to be strongly correlated?

<p><b>Observation:</b></p>
<ul>
  <li><b>Flipper length vs body mass</b>: strongly positive correlation (~0.87).</li>
  <li><b>Bill length vs bill depth</b>: moderate correlation (~0.53).</li>
  <li>Other correlations are weaker.</li>
<ul>

```python
# Load tips dataset
tips = sns.load_dataset("tips")

# Plot 1: Total bill vs tip (trend)
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.title("Total Bill vs Tip")
plt.show()

# Plot 2: Average tip by day (trend)
sns.boxplot(data=tips, x="day", y="tip", palette="Set3")
plt.title("Tip Distribution by Day")
plt.show()

# Plot 3: Histogram to find outliers
sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("Distribution of Total Bill")
plt.show()
```

<p><b>Observations / Summary:</b></p>
<ul>
  <li><b>Scatterplot:</b> Higher bills usually lead to higher tips. Friday and Saturday show slightly higher tipping trends.</li>
  <li><b>Boxplot:</b> Some tip outliers, mostly on Sunday.</li>
  <li><b>Histogram:</b> Most bills between $10â€“30, a few extreme bills above $40.</li>
