```python
import random

# Create a list of 1 million random numbers
random_numbers = [random.randint(0, 1_000_000) for _ in range(1_000_000)]

# Measure how long it takes to sort the list
%time sorted(random_numbers)
```

```python
numbers = list(range(1_000))

# List comprehension
%timeit [x**2 for x in numbers]

# For loop
def square_loop(nums):
    result = []
    for x in nums:
        result.append(x**2)
    return result

%timeit square_loop(numbers)
```

```python
%%time
total = 0
for i in range(1_000_000):
    total += i
total
```

```python
import numpy as np

numbers = np.arange(1_000_000)

# Python loop
def python_loop(nums):
    result = []
    for x in nums:
        result.append(x**2)
    return result

# List comprehension
def list_comp(nums):
    return [x**2 for x in nums]

# NumPy vectorized operation
def numpy_square(nums):
    return nums**2

# Measure runtime
%timeit python_loop(numbers)
%timeit list_comp(numbers)
%timeit numpy_square(numbers)
```
