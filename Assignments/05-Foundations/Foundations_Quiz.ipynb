## Q4. File I/O
Write Python code that saves `Hello World` to a file called `hello.txt`, then reads it back and prints the contents.

```python
# Write to file
with open("hello.txt", "w") as f:
    f.write("Hello World")

# Read from file
with open("hello.txt", "r") as f:
    content = f.read()
print(content)
```

## Q6. Plotting
Create a line plot of $y = x^2$ for $x$ values from 0 to 10.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = x**2

plt.plot(x, y)
plt.title("Line plot of y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## Q9. Timing
Compare the runtime of squaring numbers using a Python loop vs. NumPy. Use `%timeit` for each.

```python
import numpy as np

numbers = np.arange(1_000_000)

# Python loop
def python_loop(nums):
    result = []
    for x in nums:
        result.append(x**2)
    return result

# NumPy vectorized
def numpy_square(nums):
    return nums**2

# Compare runtimes
%timeit python_loop(numbers)
%timeit numpy_square(numbers)
```
