```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label="sin(x)")
plt.plot(x, y_cos, label="cos(x)", color="red")  # khác màu
plt.title("sin(x) and cos(x)")
plt.xlabel("x")
plt.ylabel("value")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.random.rand(50)
y1 = np.random.rand(50)

x2 = np.random.rand(50)
y2 = np.random.rand(50)

plt.scatter(x1, y1, label="Set 1", color="blue")
plt.scatter(x2, y2, label="Set 2", color="red")

plt.title("Two Random Scatter Sets")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ["Pizza", "Pasta", "Sushi", "Tacos"]
values = [8, 6, 9, 7]

plt.bar(categories, values)
plt.title("Favorite Foods")
plt.xlabel("Category")
plt.ylabel("Rating")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.plot(x, y, label="sin(x)")
plt.title("Customized Plot Example")
plt.xlabel("x (radians)")
plt.ylabel("sin(x)")
plt.legend()
plt.grid(True)
plt.show()
```
