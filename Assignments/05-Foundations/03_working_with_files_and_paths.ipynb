##2. Writing and Reading Text Files

```python
# Writing your name to me.txt
with open("me.txt", "w") as f:
    f.write("Nguyen Ngoc Hoang\n")

# Reading from me.txt and printing the result
with open("me.txt", "r") as f:
    contents = f.read()

print(contents)
```

##3. JSON Files

```python
import json

# Create dictionary with my name and favorite courses
data = {
    "name": "Nguyen Ngoc Hoang",
    "favorite_courses": ["Python Programming", "Data Science", "Web Development"]
}

# Write JSON to courses.json
with open("courses.json", "w") as f:
    json.dump(data, f, indent=4)  # indent=4 makes it more readable

# Read JSON back from courses.json
with open("courses.json", "r") as f:
    loaded_data = json.load(f)

loaded_data
```

```python
import csv

# Writing CSV with hobbies data
rows = [
    ["Name", "Hobby", "Skill Level"],
    ["Alex", "Photography", "Intermediate"],
    ["Sarah", "Cooking", "Beginner"],
    ["Mike", "Gardening", "Advanced"]
]

with open("hobbies.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(rows)

# Reading CSV and displaying rows
print("Contents of hobbies.csv:")
with open("hobbies.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```
